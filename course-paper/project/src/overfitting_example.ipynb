```python
import numpy as np
import matplotlib.pyplot as plt

# create a simple dataset with noise
np.random.seed(42)
x = np.linspace(0, 1, num=20)
y = np.sin(2*np.pi*x) + 0.2*np.random.randn(20)

# fit a polynomial regression model to the data
deg = 15
coeffs = np.polyfit(x, y, deg)
p = np.poly1d(coeffs)

# generate new data
x_new = np.linspace(0, 1, num=100)
y_new = np.sin(2*np.pi*x_new) + 0.2*np.random.randn(100)

# fit a polynomial regression model to the new data
deg = 15
coeffs = np.polyfit(x_new, y_new, deg)
p_new = np.poly1d(coeffs)

# plot the data and the models
plt.plot(x, y, 'ro', label='training data')
plt.plot(x_new, y_new, 'bo', label='new data')
plt.plot(x_new, p(x_new), label='model for training data')
plt.plot(x_new, p_new(x_new), label='model for new data')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# create a simple dataset with noise
np.random.seed(42)
x = np.linspace(0, 1, num=20)
y = np.sin(2*np.pi*x) + 0.2*np.random.randn(20)

# split the data into a training set and a test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# fit a polynomial regression model to the training data
deg = 15
coeffs = np.polyfit(x_train, y_train, deg)
p_train = np.poly1d(coeffs)

# evaluate the model on the test set
mse_test = np.mean((p_train(x_test) - y_test)**2)

# generate new data
x_new = np.linspace(0, 1, num=100)
y_new = np.sin(2*np.pi*x_new) + 0.2*np.random.randn(100)

# fit a polynomial regression model to the new data
deg = 15
coeffs = np.polyfit(x_new, y_new, deg)
p_new = np.poly1d(coeffs)

# plot the data and the models
plt.plot(x_train, y_train, 'ro', label='training data')
plt.plot(x_test, y_test, 'bo', label='test data')
plt.plot(x_new, p_train(x_new), label='model for training data')
plt.plot(x_new, p_new(x_new), label='model for new data')
plt.legend()
plt.show()

print('Mean squared error on test set:', mse_test)
```

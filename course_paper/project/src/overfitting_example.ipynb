```python
import numpy as np
import matplotlib.pyplot as plt

# create a simple dataset with noise
np.random.seed(42)
x = np.linspace(0, 1, num=20)
y = np.sin(2*np.pi*x) + 0.2*np.random.randn(20)

# fit a polynomial regression model to the data
deg = 15
coeffs = np.polyfit(x, y, deg)
p = np.poly1d(coeffs)

# generate new data
x_new = np.linspace(0, 1, num=20)
y_new = np.sin(2*np.pi*x_new) + 0.2*np.random.randn(20)

# fit a polynomial regression model to the new data
deg = 15
coeffs = np.polyfit(x_new, y_new, deg)
p_new = np.poly1d(coeffs)

# plot the data and the models
plt.plot(x, y, 'ro', label='training data')
plt.plot(x_new, y_new, 'bo', label='new data')
plt.plot(x_new, p(x_new), label='model for training data')
plt.plot(x_new, p_new(x_new), label='model for new data')
plt.legend()
plt.show()
```

```python
# generate new data
x_new = np.linspace(0, 1, num=100)
y_new = np.sin(2*np.pi*x_new) + 0.25*np.random.randn(100)

# plot the data and the model
plt.plot(x, y, 'ro', label='data')
plt.plot(x_new, y_new, label='unseen data')
plt.plot(x_new, p(x_new), label='model')
plt.legend()
plt.show()
```
